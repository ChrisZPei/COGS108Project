### Data Cleaning

First, we imported the relevant packages we needed to read and wrangle the dataset.

```python
!pip install xlrd
```

```python
import pandas as pd
import numpy as np
import xlrd
import warnings
warnings.filterwarnings('ignore')
```

```python
book = xlrd.open_workbook(filename='Dispatchers_Background_Data.xls')

df = pd.read_excel(book)
```

As the last two rows in the dataset do not contain relevant information, we drop them from our dataframe

```python
#remove the last two rows
df = df.iloc[:-2]
df
```

We grouped together the columns in the dataframe that pertain to stress.

```python
stress_cols = ['Mentally_Drained',
               'Phys_Drained',
               'Emergencies',
               'Lack_of_control',
               'Sleep_loss',
               'Coord_other_depts',
               'Job_pressure',
               'Scanty_Rules',
              'Mgmt_policies',
              'Job_Security',
              'Surges_in_work',
              'Communication',
              'Inadeq_Staff',
              'Resp_for_others_safety',
              'BreakTime',
              'TimeOff']
```

We dropped the rows with null values in the stress columns we defined above.

```python
#Drop the rows with NaN values
df = df.dropna(subset=stress_cols)
df
```

We added the scores from the stress columns for each of the participants and created a new column denoting the total stress value.

```python
# sum up all stress cols
df['total_stress'] = df[stress_cols].sum(axis=1)
unnessary_cols = stress_cols + ['ID',
                                'Total_years_dispatcher',
                                'Caff_Beverages',
                                'If_yes_how_many_daily',
                                'Older_Younger_or_Same',
                                'Relief_schedule',
                                'Mentally_Drained',
                                'Phys_Drained',
                                'FRA_report',
                               'SunStart_Time',
                               'MonStart_Time',
                               'TuesStart_Time',
                               'WedStart_Time',
                               'ThuStart_Time',
                               'FriStart_Time',
                               'SatStart_Time',
                               'SunEnd_Time',
                               'MonEnd_Time',
                               'TuesEnd_Time',
                               'WedEnd_Time',
                               'ThuEnd_Time',
                               'FriEnd_Time',
                               'SatEnd_Time']
# drop unnecessary cols by intuition
df = df.drop(columns=unnessary_cols)
df
```

Since the original dataset used numbers to represent answer options (e.g. For sex, 1 is male and 2 is female), we changed these numeric values to the string values that they represent to make the information easier to read.

```python
df['Sex'] = df['Sex'].replace({1:"male",2:"female"})
```

```python
df['Job_type'] = df['Job_type'].replace({1:'trick', 2:"asst chief"})
df['Diagnosed_Sleep_disorder'] = df['Diagnosed_Sleep_disorder'].fillna(2)
df['Diagnosed_Sleep_disorder'] = df['Diagnosed_Sleep_disorder'].replace({1:'Yes',2:'No'})

df['Medical_treatment'] = df['Medical_treatment'].fillna(0)
df['Medical_treatment'] = df['Medical_treatment'].replace({0:"Not Apply",1:"Yes",2:"No"})

df['Marital_Status'] = df['Marital_Status'].replace({1:"Single",2:"Married"})

df['Sleep_Apnea'] = df['Sleep_Apnea'].replace({1:"Yes",2:"No"})


df['Childrendependents'] = df['Childrendependents'].fillna(0)
df['Children_under_2_yrs'] =df['Children_under_2_yrs'].fillna(0)

df.describe()
```

We used the US average median of 40 hours worked per week, and subtracted this from the participants' reported average hours worked per week to calculate how many hours they overworked on average.

```python
df = df.dropna(subset=['Avg_Work_Hrs_Week'])
df['number of over work'] = df['Avg_Work_Hrs_Week'] -40
df = df.drop(columns=['Avg_Work_Hrs_Week'])
df.head()
```

```python
df = df.dropna(subset=['Sick_Days_in_last_year','Health_status','Total_years_present_job'])
df.head()
```

# Ethics & Privacy

We think there might be a bias in our data because most of the time people are not willing to share their weakness, so the data that we collect might be very positive from the people with less stress. Many of the survey questions on levels of stress are self-reported, which can be subject to bias and may not reflect reality. The data also will have a regional bias because we are getting data from one area’s railroad work. The different areas might produce a different data set. There should not be any issues with privacy, because the questions are answered anonymously and the ages are listed as ranges not specifics. Perhaps there can be outliers who are able to be identified within the datasets due to some stand out values, but we will likely not be using charts that present them as such unfairly.

The dataset we are using focuses on railroad employees, which is a very small subset of the working class and is not representative of the larger, more general population. The results may be biased due to the specific labor that is unique to railroad work. Additionally, there may have been certain biases amongst the groups of people who work these jobs, such as more of certain age ranges and less of others, which can lead to a biased point of view towards/against this group.

Comparing against other datasets will definitely help us with focusing too much on a certain job. Also, we can check to see how many people make up differing groups and correct for that when it comes up. Making sure to take averages across differing groups will help to mitigate any biases created by an overabundance of certain age ranges, certain years worked, etc.
