# Hypothesis


Sleeplessness: Average sleep duration(found by comparing start and end times of each day worked and taking the average) minus the hours spent sleeping of the individual.

Life events: Total number of life events. (life events include personal illness, marital woes, birth of a child, death of a spouse, change in sleep habits, trouble with law, family or friend illness, financial woes, home change, social change, family death)

Overworking: Hours worked minus national standard (40).

Illness: Sick days for the year minus the average sick days.

Guardianship: Number of children dependents.

We will make use of descriptive statistics to find out the general characteristics of our population, then use that information for correlation and regression analysis. Additionally, we can use ANOVA to compare different levels of stress within specific variables, like looking at the top half of overworked individuals and the bottom half to see if they have a significantly different average level of stress.

We believe that among the examined variables, sleeplessness will demonstrate the greatest relationship with increased levels of stress, more so than life events, illness, guardianship, or any other evaluated factors.

##### Null Hypothesis (H0): There is no significant difference in relationship between stress levels and any of the variables tested.

##### Alternative Hypothesis (H1): There is a more significant relationship between stress levels and sleeplessness than the other variables tested.

The main variables cover start and end times for work, average number of hours worked, and self-reported measures of fatigue and stress. These variables ask about feelings of stress regarding topics like sleep loss, lack of control, workplace stress, and more. The main data types in this study are time and integers, both being numeric. To answer our question, we would need to calculate the hours worked for each day based on the reported start and end times. For questions with binary responses, for example the question for gender where 1 denotes Male and 2 denotes Female, we would change these values to the strings they represent for simplicity. Since there are multiple variables covering different areas of stress, we would need to combine these results and average them out to get a total average stress score. Lastly, we would drop columns that are unnecessary to our question, like Participant ID. 
The dataset we find is in the xls format. In order to get the data into a usable format, we first import the xlrd package from the library to read the data into pandas dataframe. Since we are studying how work schedules and sleep patterns correlate to fatigue and stress, we will drop the rows with na values in columns that are related to the stress, such as Mentally_Drained, Phys_Drained, etc. After that, we calculate the total stress value by just adding up all values in the columns that relate with stress. After that we drop the columns we do not want such as Id, Start_time, and etc. Then , we change some columns with numerical objects to string objects to make our data much easier to understand. For example, we change 1 to male and 2 to female in the sex column. Next, we substract calculate the number of hours overwork per week by substracting 40 from the actual number of hours in work in a week. Finally, we check the dataframe’s value in every cell to make sure there is no NaN value.

```python
import pandas as pd
import numpy as np
import xlrd
import warnings
warnings.filterwarnings('ignore')
```

```python
book = xlrd.open_workbook(filename='Dispatchers_Background_Data.xls')

df = pd.read_excel(book)
```

```python
#remove the last two rows
df = df.iloc[:-2]
df
```

```python
stress_cols = ['Mentally_Drained',
               'Phys_Drained',
               'Emergencies',
               'Lack_of_control',
               'Sleep_loss',
               'Coord_other_depts',
               'Job_pressure',
               'Scanty_Rules',
              'Mgmt_policies',
              'Job_Security',
              'Surges_in_work',
              'Communication',
              'Inadeq_Staff',
              'Resp_for_others_safety',
              'BreakTime',
              'TimeOff']
```

```python
#Drop the rows with NaN values
df = df.dropna(subset=stress_cols)
df
```

```python
# sum up all stress cols
df['total_stress'] = df[stress_cols].sum(axis=1)
unnessary_cols = stress_cols + ['ID',
                                'Total_years_dispatcher',
                                'Caff_Beverages',
                                'If_yes_how_many_daily',
                                'Older_Younger_or_Same',
                                'Relief_schedule',
                                'Mentally_Drained',
                                'Phys_Drained',
                                'FRA_report',
                               'SunStart_Time',
                               'MonStart_Time',
                               'TuesStart_Time',
                               'WedStart_Time',
                               'ThuStart_Time',
                               'FriStart_Time',
                               'SatStart_Time',
                               'SunEnd_Time',
                               'MonEnd_Time',
                               'TuesEnd_Time',
                               'WedEnd_Time',
                               'ThuEnd_Time',
                               'FriEnd_Time',
                               'SatEnd_Time']
# drop unnecessary cols by intuition
df = df.drop(columns=unnessary_cols)
df
```

```python
df['Sex'] = df['Sex'].replace({1:"male",2:"female"})
```

```python
df['Job_type'] = df['Job_type'].replace({1:'trick', 2:"asst chief"})
df['Diagnosed_Sleep_disorder'] = df['Diagnosed_Sleep_disorder'].fillna(2)
df['Diagnosed_Sleep_disorder'] = df['Diagnosed_Sleep_disorder'].replace({1:'Yes',0:'No'})

df['Medical_treatment'] = df['Medical_treatment'].fillna(0)
df['Medical_treatment'] = df['Medical_treatment'].replace({0:"Not Apply",1:"Yes",2:"No"})

df['Marital_Status'] = df['Marital_Status'].replace({1:"Single",2:"Married"})

df['Sleep_Apnea'] = df['Sleep_Apnea'].replace({1:"Yes",2:"No"})


df['Childrendependents'] = df['Childrendependents'].fillna(0)
df['Children_under_2_yrs'] =df['Children_under_2_yrs'].fillna(0)

df.describe()
```

```python

df = df.dropna(subset=['Avg_Work_Hrs_Week'])
df['number of over work'] = df['Avg_Work_Hrs_Week'] -40
df = df.drop(columns=['Avg_Work_Hrs_Week'])
df.head()
```

```python
df = df.dropna(subset=['Sick_Days_in_last_year','Health_status','Total_years_present_job'])
df
```

# Ethics & Privacy

We think there might be a bias in our data because most of the time people are not willing to share their weakness, so the data that we collect might be very positive from the people with less stress. Many of the survey questions on levels of stress are self-reported, which can be subject to bias and may not reflect reality. The data also will have a regional bias because we are getting data from one area’s railroad work. The different areas might produce a different data set. There should not be any issues with privacy, because the questions are answered anonymously and the ages are listed as ranges not specifics. Perhaps there can be outliers who are able to be identified within the datasets due to some stand out values, but we will likely not be using charts that present them as such unfairly.

The dataset we are using focuses on railroad employees, which is a very small subset of the working class and is not representative of the larger, more general population. The results may be biased due to the specific labor that is unique to railroad work. Additionally, there may have been certain biases amongst the groups of people who work these jobs, such as more of certain age ranges and less of others, which can lead to a biased point of view towards/against this group.

Comparing against other datasets will definitely help us with focusing too much on a certain job. Also, we can check to see how many people make up differing groups and correct for that when it comes up. Making sure to take averages across differing groups will help to mitigate any biases created by an overabundance of certain age ranges, certain years worked, etc.
