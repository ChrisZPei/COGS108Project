#### Calculating Stress

We grouped together the columns in the dataframe that pertain to stress. We dropped any null values so they do not interfere with our testing..

```python
stress_cols = ['Mentally_Drained',
               'Phys_Drained',
               'Emergencies',
               'Lack_of_control',
               'Sleep_loss',
               'Coord_other_depts',
               'Job_pressure',
               'Scanty_Rules',
              'Mgmt_policies',
              'Job_Security',
              'Surges_in_work',
              'Communication',
              'Inadeq_Staff',
              'Resp_for_others_safety',
              'BreakTime',
              'TimeOff']
```

```python
#Drop the rows with NaN values
df = df.dropna(subset=stress_cols)
df.shape
```

We are now working with 437 responses.

We added the scores from the stress columns for each of the participants and created a new column denoting the total stress value. We then kept only the columns pertaining to the variables we are exploring, and dropped all other columns.

```python
# sum up all stress cols
df['total_stress'] = df[stress_cols].sum(axis=1)

#keep relevant columns only
df = df[['Childrendependents', 'Sick_Days_in_last_year', 'Diagnosed_Sleep_disorder', 
         'Avg_Work_Hrs_Week', 'Total_life_events', 'total_stress']]
df.head()
```

#### Standardizing Binary Responses

Participants responded with a 1 if they did have a diagnosed sleep disorder, and a 2 if not. We chagned these responses from the numeric values to the string values they represent for simplicity.

```python
df['Diagnosed_Sleep_disorder'] = df['Diagnosed_Sleep_disorder'].replace({1:'Yes',2:'No'})
df.head()
```

#### Handling Null Values

Based on our understanding, we felt that null values for certain questions correspond to Not Applicable answers. For diagnosed sleep disorder, we chose to take nulls as having no sleep disorders. For sick days taken in the last year, we took null values as 0 days taken. For children dependents, we took nulls as 0 dependents.

```python
df['Diagnosed_Sleep_disorder'] = df['Diagnosed_Sleep_disorder'].fillna('No')
df['Sick_Days_in_last_year'] = df['Sick_Days_in_last_year'].fillna(0)
df['Childrendependents'] = df['Childrendependents'].fillna(0)
```

We then checked how many remaining null values there were.

```python
df.isna().sum()
```

For average work hours per week, we felt that null values could pertain to other scenarios so we decided to not fill them, and instead dropped the rows so they would not interfere with our testing.

```python
df = df.dropna(subset='Avg_Work_Hrs_Week')
```

```python
df[df.isnull().any(axis=1)]
```

```python
df['Childrendependents'] = df['Childrendependents'].round().astype(int)
df['Sick_Days_in_last_year'] = df['Sick_Days_in_last_year'].round().astype(int)
df['Avg_Work_Hrs_Week'] = df['Avg_Work_Hrs_Week'].round().astype(int)
df['Total_life_events'] = df['Total_life_events'].round().astype(int)
df['total_stress'] = df['total_stress'].round().astype(int)
```

## Part II: Looking at the Statistics

Though the plots we created help us visualize the data, we cannot determine relationships and effect sizes based off of the plots alone. We will conduct various statistical tests to quantify our findings and explore relationships on a deeper level.

### Setup: Importing Testing Packages

```python
# Statmodels & patsy
import patsy
import statsmodels.api as sm
from patsy import dmatrices
from statsmodels.stats.outliers_influence import variance_inflation_factor

import warnings
warnings.filterwarnings('ignore')

from scipy.stats import ttest_ind
```

### Pearson's Correlation

```python
df.corr(numeric_only=True)
```

### Multiple Colinearity

Before we can move on to linear regression, we must ensure that there is no multiple colinearity, that is, our independent variables are not strongly associated with one another so as to not impact our future tests. To test for multiple colinearity, we calculated the variance inflation factor (VIF) because we have five independent variables.

```python
#create design matrix with IVs
y, X = dmatrices('total_stress ~ Diagnosed_Sleep_disorder + Total_life_events + Avg_Work_Hrs_Week + Sick_Days_in_last_year + Childrendependents', data=df, return_type='dataframe')

#create new dataframe to hold VIF
vif_df = pd.DataFrame()
vif_df['variable'] = X.columns 

#calculate VIF for each IV 
vif_df['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
 
print(vif_df)
```

As we can see, all of the VIF values are close to 1, meaning that there is no correlation between the independent variables. We are ready to move on to linear regression.

### Multiple Linear Regression

We will use multiple linear regression to measure the effect of our five independent variables on our outcome of total stress, and use the p-values to identify which relationships are significant. We will use an alpha value of 0.05.

```python
outcome, predictors = patsy.dmatrices('total_stress ~ Diagnosed_Sleep_disorder + Total_life_events + Avg_Work_Hrs_Week + Sick_Days_in_last_year + Childrendependents', df)
mod = sm.OLS(outcome, predictors)
res = mod.fit()
print(res.summary())
```

From these results, we can see that only diagnosed sleep disorder and total life events resulted in significant p-values less than 0.05. Looking at the effect sizes, having a diagnosed sleep disorder had a larger effect size at 3.83 compared to the effect size of 1.31 for life events.

### Weighted Linear Regression
 Since the number of railroad workers who do not have sleep disorder is around 10 times than the workers who have sleep disorder. We want to consider increasing the weights of the instances which answering "Yes" to the "Diagnosed_Sleep_disorder" by 10 times. 

```python
df["weights"] = df['weights'] = df['Diagnosed_Sleep_disorder'].apply(lambda x: 10 if x == 'Yes' else 1)
weights = df['weights']
outcome, predictors = patsy.dmatrices('total_stress ~ Diagnosed_Sleep_disorder + Total_life_events + Avg_Work_Hrs_Week + Sick_Days_in_last_year + Childrendependents', df)
mod = sm.WLS(outcome, predictors,weights=weights)
res = mod.fit()
print(res.summary())
```

<font size="3">Based on the results from the weighted regression, we can see that only diagnosed sleep disorder and total life events resulted in significant p-values less than 0.05. Looking at the effect sizes, having a diagnosed sleep disorder had a larger effect size at 4.09 compared to the effect size of 1.56 for life events. The effect sizes from the weighted linear regression are similar to the ordinary linear regression. </font>

To answer these questions, we first created visualizations to explore the distribution of our data through histograms, and used scatter plots to visualize the relationships between our independent variables and stress. We then conducted further statistical tests to determine which relationships were significant, and to measure the effect sizes of these variables.

For our question on whether there are relationships between any of our independent variables and stress, we reject the null hypothesis and conclude that sleeplessness and life events are both related to stress for railroad workers. To come to this conclusion, we first conducted a test for multiple collinearity by calculating the variance inflation factor value to ensure that our independent variables were not highly correlated with each other. Once we were sure there was no multiple collinearity, we moved on to conducting the multiple linear regression test. Through this, we were able to see which variables had a significant relationship with stress. Only sleeplessness and life events had a p-value less than our alpha value of 0.05, so we can conclude that these are significant.

For our question regarding which variables have the largest effect on stress, we reject the null hypothesis and conclude that sleeplessness had the largest effect size on railroad workers' stress. To determine this, we looked back at our multiple regression test. Between our two significant variables, sleeplessness and life events, sleeplessness had an effect size of 3.83, and life events had an effect size of 1.31 in ordinary least square regression; also, having a effect size of 4.09 and a effect size of 1.56 in weighted least square regression for these two variable respectively. From this, we conclude that sleeplessness had the largest impact on stress.
